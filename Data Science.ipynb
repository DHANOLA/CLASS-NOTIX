# reading the dataset

```python
import numpy as np
import pandas as pd

df = pd.read_csv("coursera_dataset.csv")
df.head()
```

# removing null  values

```python
# replace "?" by "Nan"
df = df[df.columns].replace("?", np.nan)
df
```

```python
# drop the NaN values

df = df.dropna(subset=df.columns, axis=0)
df
```

# changing the datatypes

```python
# first analysisof  the datatypes of each column

df.dtypes
```

```python
# getting the column names with incorect dtype

col = list()
index = [1, 13, 18, 19, 21, 22, 25]
for i in range(len(df.columns)):
    if i in index:
        col.append(df.columns[i])
col
```

```python
# creating the list with  correct datatype for each column

typp = [np.int64, np.float64, np.float64,
        np.float64, np.int64, np.int64, np.int64]
```

```python
# now setting the new dtype tpo the columns

for i in range(len(col)):
    df[col[i]] = df[col[i]].astype(typp[i])
```

```python
# again analysis of  the datatypes of each column after the modification

df.dtypes
```
