#### 2.4 HISTOGRAMS PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n={}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc=0.0, scale=1.0, size=sample_size)
    axs[n].hist(sample, bins=100)
    axs[n].set_title('n={}'.format(sample_size))
```

#### 2.5 INTERACTIVITY PLOT

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

plt.figure()
data = np.random.randn(10)
plt.plot(data)


def onclick(event):
    plt.cla()
    plt.plot(data)
    plt.gca().set_title('Event at pixels {}, {}, and data {}, {}'.format(
        event.x, event.y, '\n', event.xdata, event.ydata))


plt.gcf().canvas.mpl_connect('button_press_event', onclick)
```

```python
import pandas as pd
from random import shuffle

origins = [
    'China', 'Brazil', 'India', 'USA', 'Canada', 'UK', 'Germany', 'Iraq',
    'Chile', 'Mexico'
]
shuffle(origins)

df = pd.DataFrame({
    'height': np.random.rand(10),
    'weight': np.random.rand(10),
    'Origin': origins
})
df
```

```python
plt.figure()
plt.scatter(df['height'], df['weight'], picker=5)
plt.gca().set_ylabel('Weight')
plt.gca().set_xlabel('Height')
```

```python
def onpick(event):
    origins = df.iloc[event.ind[0]]['origin']
    plt.gca().set_title('Selected item came from {}'.format(origin))


plt.gcf().canvas.mpl_connect('picl_event', onpick)
```

### 5 SEABORN LIBRARY

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 1000), name='v1')
v2 = pd.Series(2 * v1 + np.random.normal(60, 15, 1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50, 150, 5), label='v1')
plt.hist(v2, alpha=0.7, bins=np.arange(-50, 150, 5), label='v2')
plt.legend()
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', density=True)
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'})
```

```python
sns.jointplot(v1, v2, alpha=0.4)
```

```python
grid = sns.jointplot(v1, v2, alpha=0.4)
grid.ax_joint.set_aspect('equal')
```

```python
plt.figure()
sns.jointplot(v1, v2, kind='hex')

sns.set_style('white')
sns.jointplot(v1, v2, kind='kde', space=0)
```

### 6 SUBPLOT

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
plt.figure()

plt.subplot(1, 2, 1)
linear_data = np.array([1, 2, 3, 4, 5, 6, 7, 8])

plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2

plt.subplot(1, 2, 2)
plt.plot(exponential_data, '-o')
```

```python
plt.subplot(1, 2, 1)
plt.plot(exponential_data, '-x')
```

```python
# To lock the y-axis for two plots side-by-side
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')
ax2 = plt.subplot(1, 2, 2, sharey=ax1)
plt.plot(exponential_data, '-x')
```
